```python
X_data=data.drop(['Credit-Application-Result'],axis=1)
y_data=data['Credit-Application-Result']


y_data = y_data.apply(lambda x: 0 if x == 'Non-Creditworthy' else 1)

columnsToEncode=list(X_data.select_dtypes(include=['object']))
X_data= pd.get_dummies(X_data, columns=columnsToEncode)

columnsToEncode=['Credit-Application-Result']
#y_data= pd.get_dummies(y_data, columns=columnsToEncode)

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X_data,y_data,test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV,ShuffleSplit,KFold
from sklearn.metrics import make_scorer,fbeta_score,accuracy_score

def fit_DT_model(X,y):
    cv_set=ShuffleSplit(n_splits=10,test_size = 0.20,random_state=42)
    #cv_set=KFold(n_splits=10,random_state=42)
    
    params={'max_depth':range(1,11),'splitter':['best','random'],'min_samples_split':range(2,20) }
    score_func=make_scorer(accuracy_score)
    classifier=DecisionTreeClassifier()
    grid=GridSearchCV(classifier,params,score_func,cv=cv_set)
    grid=grid.fit(X,y)
    print ("Parameters for the optimal model.",grid.best_params_)
    return grid.best_estimator_

def fit_model(classifier,params,X,y):
    cv_set=ShuffleSplit(n_splits=10,test_size = 0.20,random_state=42)
    score_func=make_scorer(accuracy_score)
    grid=GridSearchCV(classifier,params,score_func,cv=cv_set)
    grid=grid.fit(X,y)
    print ("Parameters for the optimal model.",grid.best_params_)
    return grid.best_estimator_
```

```python
from sklearn.metrics import classification_report

def get_score(classifier,X,y):
    y_pred=classifier.predict(X)
    score=accuracy_score(y,y_pred)
    print('accuracy score',score)
    print(classification_report(y,y_pred))
    
def get_all_scores(clf,X_train,y_train,X_test,y_test):
    print("train dataset")
    get_score(clf,X_train,y_train)
    
    print("test dataset")
    get_score(clf,X_test,y_test)
    
def feature_plot(imp_features,X):
    indices = np.argsort(imp_features)[::-1]
    num_features=len(imp_features[imp_features>0])
    columns = X.columns.values[indices][:num_features]
    values=imp_features[indices][:num_features]
    
    #for f in range(num_features):
    #    print(columns[f], imp_features[indices[f]])
    plt.figure(figsize = (15,5))
    plt.title("Feature importances")
    plt.barh(range(num_features), values, align="center")
    plt.yticks(range(num_features), columns)
    plt.ylim([ num_features,-1])
    plt.show()    
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB


#view basic models
DT_clf=DecisionTreeClassifier()
RF_clf=RandomForestClassifier()
SGD_clf=SGDClassifier()
MLP_clf=MLPClassifier()
AB_clf=AdaBoostClassifier()
KN_clf=KNeighborsClassifier()
GB_clf=GradientBoostingClassifier()
SVC_clf=SVC()
NB_clf=GaussianNB()
for clf in [DT_clf,RF_clf,SGD_clf,MLP_clf,AB_clf,KN_clf,GB_clf,SVC_clf,NB_clf]:
    print('classifier: ',clf.__class__.__name__)

    clf_model=clf.fit(X_train,y_train)
   # get_all_scores(clf_model,X_train,y_train,X_test,y_test)
    print("test dataset")
    get_score(clf_model,X_test,y_test)

```

```python
 #TOP 3 MODELS with GridSearchCV 
params={'max_depth':[1,8,12,20,30],'splitter':['best','random'],'min_samples_split':[2,5,10,20] }
DT_clf=fit_model(DT_clf,params,X_train,y_train)
get_score(DT_clf,X_test,y_test) 

#Random Forest
params={'criterion':['gini','entropy'],'n_estimators':[10,50,100,200,500,1000],'max_depth':[1,8,12,20,30],'min_samples_split':[2,5,10,20]}
RF_clf=fit_model(RF_clf,params,X_test,y_test)
get_score(RF_clf,X_test,y_test)

#Gradient Boost
params={'loss':['exponential','deviance'],'learning_rate':[0.001,0.01,0.1,1],'n_estimators':[100,200,500,1000,5000],'max_depth':[1,8,12,20,30],'min_samples_split':[2,5,10,20]}
GB_clf=fit_model(GB_clf,params,X_test,y_test)
get_score(GB_clf,X_test,y_test)

#Adaboost
params={'n_estimators':[100,200,500,1000,5000],'learning_rate':[0.001,0.01,0.1,1]}
AB_clf=fit_model(AB_clf,params,X_test,y_test)
get_score(AB_clf,X_test,y_test)
```
